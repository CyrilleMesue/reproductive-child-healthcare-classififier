# Separating Input Features and Output Features
Before building any machine learning model, we always separate the input variables and output variables. Input variables are those quantities whose values are changed naturally in an experiment, whereas output variable is the one whose values are dependent on the input variables. So, input variables are also known as independent variables as its values are not dependent on any other quantity, and output variable/s are also known as dependent variables as its values are dependent on other variable i.e. input variables. Like here in this data, we want to predict whether the meteor is a threat to the Earth or not, so the variable **Hazardous** is our target variable and remaining features are input variable.

By convention input variables are represented with 'X' and output variables are represented with 'y'.

```python
X = df.drop('fetal_health', axis = 1)   # here we are dropping the Target feature as this is the target and 'X' is input features, the changes are not 
                                              # made inplace as we have not used 'inplace = True'
y = df['fetal_health']             # Output/Dependent variable
```

# Splitting the data into Train and Test Set
We want to check the performance of the model that we built. For this purpose, we always split (both input and output data) the given data into training set which will be used to train the model, and test set which will be used to check how accurately the model is predicting outcomes.

For this purpose we have a class called 'train_test_split' in the 'sklearn.model_selection' module.


```python
def transform_data(X, y = None):
    """
    function helps to transform data into a form that Pipeline takes
    """
    X = pd.DataFrame(X)
    new_X = []
    new_y = []
    for i in range(len(X)):
        data = {}
        for key in X.columns:
            data[key] = X[key].iloc[i]
        new_X.append(data)

    # transform y if y is given
    if type(y) != type(None):
        for i in range(len(y)):
            if y[i] == 0:
                new_y.append("class1")
            elif y[i] == 1:
                new_y.append("class2")
            else:
                new_y.append("class3")
            
    return new_X, new_y

```

```python
#train test split. Its not necessary to keep a large percentage for 
#testing since cross validation will be used to evaluate model
X_train, X_test, y_train, y_test = train_test_split(np.array(X), np.array(y), test_size=0.15)

X_temp = X_train
y_temp = y_train

X_train, y_train = transform_data(X_temp, y_temp)
X_test, y_test = transform_data(X_test, y_test)
```

### Solving Class Imbalance

```python
from collections import Counter
from sklearn.utils import resample
```

```python
class_count = Counter(y_train)
min_count = min(class_count.values())
total = len(y_train)
```

```python
# The non-smoking class is approximately double the smoking class. 
# Therefore we can penalize the smoking class twice
class_weight = {"class1" : (total - class_count["class1"])/min_count,
 "class2" : (total - class_count["class2"])/min_count,
 "class3" : (total - class_count["class3"])/min_count}
```

```python
def balance_dataset(data, sampling = "upsampling", balacing_ratio = 1.0):
    
    """
    data: a dataframe containing features and labels for both classes 
    
    sampling: If set to upsampling, the the class with fewer samples 
              will be increased else, the class with more samples will be reduced
              
    balancing_ratio: Determines how many percent of the dataset sould be increased or decreased
    
    NOTE: This function works only on datasets with 2 classes
    
    returns : a dataframe with data resampled such that the classes are balanced or nearly balanaced
    """
    # unique classes 
    classes = list(set(data.labels))
    # separate data into the two classes
    
    class1 = data[data.labels == classes[1]].reset_index().drop(columns = "index")
    class0 = data[data.labels == classes[0]].reset_index().drop(columns = "index")
    
    # perform upsampling
    if sampling == "upsampling":
        
        #amount to increase sample
        increase_amount = abs(int((len(class1) - len(class0))*balacing_ratio))
        
        if len(class1) > len(class0):
            
            # amount of class0 that will be added
            extra_class0 = resample(class0, n_samples=increase_amount, replace=True, random_state=0)
            
            df = pd.concat([class0,extra_class0,class1], axis = 0)
        else:
            # amount of class1 that will be added
            extra_class1 = resample(class1, n_samples=increase_amount, replace=True, random_state=0)
            
            df = pd.concat([class0,class1,extra_class1], axis = 0)
        
        return df.reset_index().drop(columns = "index")
        
    elif sampling == "downsampling":

        #amount to increase sample
        decrease_amount = abs(int((len(class1) - len(class0))*(1-balacing_ratio)))
        
        if len(class1) > len(class0):
            
            # amount of class0 that will be added
            new_class1 = resample(class1, n_samples=len(class0)+decrease_amount, replace=False, random_state=0)
            
            df = pd.concat([class0,new_class1], axis = 0)
        else:
            # amount of class1 that will be added
            new_class0 = resample(class0, n_samples=len(class1)+decrease_amount, replace=False, random_state=0)
            
            df = pd.concat([new_class0,class1], axis = 0)
        
        return df.reset_index().drop(columns = "index")
    else:
        return data
                  

def multiclass_balancing(data, sampling = "upsampling", balacing_ratio = 1.0):
    
    """
    Performs upsampling or downsampling of multi-class dataset
    """
                  
    classes = list(set(data.labels))
    
    class_count = Counter(list(data.labels))
                  
    max_class = [cl for cl in class_count if class_count[cl] == max(class_count.values())][0]
         
    maxclass_data = data[data.labels == max_class]
    new_data = data[data.labels == max_class]     
    
    for cl in classes:

        if cl != max_class:
            class2_data = data[data.labels == cl]
            temp_data = pd.concat([maxclass_data, class2_data], axis = 0)

            resampled_data = balance_dataset(temp_data, sampling = sampling, balacing_ratio = balacing_ratio)
            resampled_data = resampled_data[resampled_data.labels == cl]
            
            new_data = pd.concat([new_data,resampled_data], axis = 0)
            
    new_data = new_data.reset_index().drop(columns = "index")
    return new_data
```

***Why do we need to do the same procedure of filling missing values, data cleaning and data preprocessing on the new test data as it was done for the training and validation data?***

**Ans:** Because our model has been trained on certain format of data and if we don't provide the testing data of the similar format, the model will give erroneous predictions and the rmse of the model will increase. Also, if the model was build on 'n' number of features, while predicting on new test data you should always give the same number of features to the model. In this case if you provide different number of features while predicting the output, your ML model will throw a ValueError saying something like 'number of features given x; expecting n'. Not confident about these statements? Well, as a data scientist you should always perform some experiment and observe the results.
